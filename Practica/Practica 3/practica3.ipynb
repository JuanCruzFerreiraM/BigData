## 1

El objetivo es resolver un sistema de 15 ecuaciones con 15 incógnitas usando el método de Jacobi. En este caso voy a seguir una implementación similar a la vista en clase. 

```python
import datetime,math,re
import random
import sys
sys.path.append('../..')

from MRE import Job


root_path = './' #Simplemente para poder usar rootpaths
input_path = root_path + "input/"
output_path = root_path + "output/"
input_dir = input_path + "eje1/"
output_dir1 = output_path + "eje1/"


def fmap1(key, value, context):
    var_values = context["unknowns"]
    coefs = value.split('\t')
    res = 0
    for i in range(15):
        res += var_values[i] * float(coefs[i])
    context.write(key, res)

def fred1(key, value, context):
    res = 0
    for v in value:
        res = v
    context.write(key,v)

def read_unknowns ():
    output_file = output_dir1 + "output.txt"
    unknowns = [1] * 16  # Inicializa con 1 en todos los elementos
    with open(output_file, "r", encoding="utf-8") as f:
        for line in f:
            var, value = line.strip().split('\t')
            idx = int(var.replace("var", ""))
            unknowns[idx] = float(value)

    unknowns[0] = 1
    return unknowns

def calc_error (unknowns0, unknowns1):
    res = 0
    for i in range(15):
        res += (unknowns1[i+1] - unknowns0[i + 1]) ** 2
    return res
    
job1 = Job(input_dir,output_dir1,fmap1,fred1)
incog = {"unknowns": [1,1,2,3,4,5,6,7,1,2,5,2,2,1,3,4]}

error = 0.01
dif  = 1
i = 0
while(dif >= error):
    i += 1
    job1.setParams(incog)
    job1.waitForCompletion()
    unknowns = read_unknowns()
    dif = calc_error(incog["unknowns"], unknowns)
    incog["unknowns"] = unknowns
print(i)

```

En este caso tenemos que solucionar el mismo problema pero de distinta forma, lo que yo estoy pensando que se puede hacer es que cada mapper, devuelva algo estilo <var_i, (var_j, value)>
De esta forma cada reducer se encarga de una de las variables y calcula la suma devolviendo algo estilo <var_i, value> el resto es igual

```python
import datetime,math,re
import random
import sys
sys.path.append('../..')

from MRE import Job


root_path = './' #Simplemente para poder usar rootpaths
input_path = root_path + "input/"
output_path = root_path + "output/"
input_dir = input_path + "eje2/"
output_dir1 = output_path + "eje2/"


def fmap1(key, value, context):
    data = value.split('\t')
    var = data[0]
    var_value = data[1]
    context.write(key, (var, var_value))
    

def fred1(key, value, context):
    res = 0
    var_value = context["unknowns"]
    for v in value: 
        coef = v[0]
        coef_value = v[1]
        if (coef == "TI"):
            idx = 0
        else: 
            idx = int(coef.replace("var", ""))
        res += var_value[idx] * float(coef_value)
    context.write(key,res)

def read_unknowns ():
    output_file = output_dir1 + "output.txt"
    unknowns = [1] * 16  # Inicializa con 1 en todos los elementos
    with open(output_file, "r", encoding="utf-8") as f:
        for line in f:
            var, value = line.strip().split('\t')
            idx = int(var.replace("var", ""))
            unknowns[idx] = float(value)

    unknowns[0] = 1
    return unknowns

def calc_error (unknowns0, unknowns1):
    res = 0
    for i in range(15):
        res += (unknowns1[i+1] - unknowns0[i + 1]) ** 2
    return res
    
job1 = Job(input_dir,output_dir1,fmap1,fred1)
incog = {"unknowns": [1,-37,8,-5,4,5,6,7,1,2,5,2,2,1,3,4]}

error = 0.01
dif  = 1
i = 0
while(dif >= error):
    i += 1
    job1.setParams(incog)
    job1.waitForCompletion()
    unknowns = read_unknowns()
    dif = calc_error(incog["unknowns"], unknowns)
    incog["unknowns"] = unknowns
print(i)

```
